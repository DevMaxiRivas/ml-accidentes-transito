# Contexto
Según la Organización Mundial de la Salud (OMS), los siniestros viales son una de las principales causas de muerte y discapacidad en el mundo, con un impacto devastador en las víctimas y sus familias, y una carga significativa para los sistemas de salud y la economía. En la República Argentina, las muertes viales representan una preocupación de salud pública urgente, con un elevado número de víctimas cada año.

El conjunto de datos sobre los hechos y víctimas de muertes viales en Argentina ofrece una oportunidad para analizar y comprender los factores que contribuyen a estos siniestros. La información contenida incluye variables como el tipo de siniestro, la ubicación, las características de las víctimas, y las condiciones del entorno, proporcionando una base sólida para el estudio de las causas y patrones de los accidentes viales.

Mediante el uso de técnicas de análisis de datos y modelos predictivos, esta información puede ayudar a identificar áreas de riesgo y a desarrollar estrategias de prevención efectivas, lo cual resulta fundamental para reducir el número de muertes y lesiones graves en las rutas y calles del país. Este recurso contribuye a una comprensión integral de la problemática de los siniestros viales, fomentando políticas públicas y esfuerzos de concientización orientados a mejorar la seguridad vial en Argentina.

# Contenido
Este conjunto de datos incluye 45.003 registros de accidentes provenientes de diferentes contextos geográficos, y proporciona información detallada sobre cada suceso.

Las características incluyen:
- id_hecho: Código de identificación del hecho.
- federal: Indica si el dato es reportado por alguna fuerza federal.
- tipo_persona: Caracteriza el tipo de persona: víctima o inculpado.
- tipo_persona_id: Código identificador de la persona involucrada en el hecho.

- provincia_nombre: Nombre de la provincia donde ocurrió el hecho.
- departamento_id: Código identificador del departamento de ocurrencia del hecho.
- departamento_nombre: Nombre del departamento geográfico donde ocurrió el hecho.
- localidad_id: Código de la localidad donde ocurrió el hecho según el Instituto Nacional de Estadística y Censo (INDEC).
- localidad_nombre: Nombre de la localidad donde ocurrió el hecho.
- latitud: Latitud estimada del hecho
- longitud: Longitud estimada del hecho
- anio: Año de ocurrencia del hecho.
- mes: Mes de ocurrencia del hecho.
- fecha_hecho: Fecha del hecho.
- hora_hecho: Hora del hecho.
- calle_nombre: Nombre de la calle o ruta donde sucedió el hecho.
- calle_altura: Altura de la calle o km de la ruta en la que ocurrió el hecho.
- calle_interseccion: Determina si el hecho ocurrió en una intersección de calles.
- calle_interseccion_nombre: Calle de intersección (que corta perpendicularmente a la registrada en la variable Calle) en la que ocurrió el hecho.
- semaforo_estado: Registra estado del funcionamiento del semáforo en el momento y el lugar en el que aconteció el hecho.
- tipo_lugar: Tipo de lugar en el que aconteció el hecho.
- modo_produccion_hecho: Registra el modo o la modalidad en la cual se produjo el hecho.
- modo_produccion_hecho_ampliada: Modo de producción del hecho con mayor cantidad de tipologías.
- modo_produccion_hecho_otro: Especificación del modo en el cual se produjo el hecho en caso de no estar incluido en las categorías predeterminadas.
- clima_condicion: Registra las condiciones climáticas presentes en el momento y el lugar del hecho.
- clima_otro: Especificación de las condiciones climáticas presentes en momento y lugar del hecho en caso de no estar incluido en las categorías predeterminadas.
- motivo_origen_registro: Registra el motivo que origina el registro del hecho.
- motivo_origen_registro_otro: Especificación del motivo que origina el registro del hecho en caso de no estar incluido en las categorías predeterminadas.

- victima_clase: Clase de la víctima del hecho.
- victima_clase_otro: Especificación de clase de víctima en caso de no estar incluido en las categorías predeterminadas de la variable Clase victima.
- victima_vehiculo: Registra el tipo de vehículo de la víctima del hecho.
- victima_vehiculo_ampliado: Tipo de vehículo de la víctima con mayor cantidad de categorías.
- victima_vehiculo_otro: Especificación del tipo de vehículo de la víctima del hecho en caso de no estar incluido en las categorías predeterminadas de la variable vehículo victima.
- victima_identidad_genero: Género de la víctima del hecho. Según Ley de Identidad de Género Nro. 26.743.
- inculpado_sexo: Sexo del inculpado del hecho.
- inculpado_tr_edad: Edad de la persona involucrada en el siniestro (en tramos).
- inculpado_vehiculo: Registra el tipo de vehículo de la inculpado del hecho.
- inculpado_vehiculo_ampliado: Tipo de vehículo del inculpado con más detalle de categorías.
- inculpado_vehiculo_otro: Especificación del tipo de vehículo del imputado del hecho en caso de no estar incluido en las categorías predeterminadas de la variable vehículo imputado.
- inculpado_genero: Género del imputado del hecho. Según Ley de Identidad de Género Nro. 26.743.

## Analizar y eliminar filas con valores faltantes en columnas críticas

```python
def mantener_filas_con_cols_criticas(df):
  columnas_criticas = [
    'latitud',
    'longitud',
    'fecha_hecho',
    'calle_interseccion',
    'semaforo_estado'
  ]

  # Contar filas con valores faltantes en columnas críticas
  filas_antes = len(df)
  df_cleaned = df.dropna(subset=columnas_criticas)
  filas_eliminadas = filas_antes - len(df_cleaned)

  print(f"\nFilas eliminadas por valores faltantes en columnas críticas: {filas_eliminadas}")
  print(f"Porcentaje de filas eliminadas: {(filas_eliminadas/filas_antes)*100:.2f}%")
  print(f"Dimensiones del dataset: {df_cleaned.shape}")

  return df_cleaned
```

# Codificación de Variables

```python
def convertir_cols(df):
    # Convertir 'fecha_hecho' a un formato de fecha
  df['fecha_hecho'] = pd.to_datetime(df['fecha_hecho'], format='%d-%m-%Y')

  # Crear la nueva columna 'dia_hecho'
  df['dia'] = df['fecha_hecho'].dt.day

  # Eliminar la columna 'fecha_hecho'
  df.drop(columns=['fecha_hecho'], inplace=True)
```

## Unión de Atributos
Una fila del dataset corresponde a una victima o inculpado por ende aquellos atributos que tienen como prefijo "victima_" si la fila corresponde a una victima tendrán datos, y todos aquellos con prefijo "inculpado_" tendran el valor "No corresponde" por ello se decidio unificar estas columnas.

```python

import pandas as pd
import numpy as np
def unificar_columnas(df):
    # Crear las nuevas columnas combinando víctima e inculpado
    df['sujeto_sexo'] = df.apply(
        lambda x: x['victima_sexo'] if x['victima_sexo'] != 'No corresponde' else x['inculpado_sexo'],
        axis=1
    )

    df['sujeto_tr_edad'] = df.apply(
        lambda x: x['victima_tr_edad'] if x['victima_tr_edad'] != 'No corresponde' else x['inculpado_tr_edad'],
        axis=1
    )


    df['sujeto_vehiculo'] = df.apply(
        lambda x: x['victima_vehiculo'] if x['victima_vehiculo'] != 'No corresponde' else x['inculpado_vehiculo'],
        axis=1
    )

    # Eliminar las columnas originales
    columnas_a_eliminar = [
        'victima_sexo', 'victima_tr_edad', 'victima_vehiculo',
        'inculpado_sexo', 'inculpado_tr_edad', 'inculpado_vehiculo'
    ]

    df_nuevo = df.drop(columns=columnas_a_eliminar)

    return df_nuevo
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def filtra_df():
  
  PATH = r"G:\Mi unidad\UNSa\4to\2DO CUATRIMESTRE\Metodos Cuantitativos para la Toma de Decisiones\TP Final\Dataset"
  df = pd.read_csv(PATH + '\SAT-MV-BU_2017-2022.csv')
  
  df = eliminar_cols_irrelevantes(df)
  df = mantener_filas_con_cols_criticas(df)
  convertir_cols(df)
  df = unificar_columnas(df)

  return df
```

## Análisis del Dataset
Como el dataset esta orientado a los sujetos involucrados en accidentes se opto por modificar el dataset de manera que se tenga presente a los accidentes como un todo en lugar de tener presente a los protagonistas. Se agregará dos columnas
"nro_inputados" y "nro_victimas". Ademas de quedarnos con aquellos accidentes los cuales se tenga determinado el motivo por el cual se produjo.

```python
def obtener_accidentes():
    df = filtra_df()
    df_accidentes = (
        df.groupby("id_hecho")
        .agg(
            {
                # Mantener información única del accidente
                "provincia_nombre": "first",
                "departamento_nombre": "first",
                "localidad_nombre": "first",
                "latitud": "first",
                "longitud": "first",
                "anio": "first",
                "mes": "first",
                "dia": "first",
                "hora_hecho": "first",
                "calle_interseccion": "first",
                "semaforo_estado": "first",
                "tipo_lugar": "first",
                "modo_produccion_hecho": "first",
                "clima_condicion": "first",
                # Contar número de sujetos por tipo_persona
                "tipo_persona": lambda x: (x == "Imputado").sum(),  # Contar imputados
            }
        )
        .rename(columns={"tipo_persona": "nro_inputados"})
    )  # Renombrar columna

    # Agregar la columna 'nro_victimas'
    df_accidentes["nro_victimas"] = df.groupby("id_hecho")["tipo_persona"].apply(
        lambda x: (x == "Víctima").sum()
    )

    # Reiniciar el índice para obtener un dataset limpio
    df_accidentes.reset_index(inplace=True)

    # Eliminar columna id_hecho
    df_accidentes = df_accidentes.drop(columns=["id_hecho"])

    # Horas despues de la media Noche
    df_accidentes["hora_hecho"] = (
        pd.to_datetime(df_accidentes["hora_hecho"], format="%H:%M:%S").dt.hour * 60
        + pd.to_datetime(df_accidentes["hora_hecho"], format="%H:%M:%S").dt.minute
    )

    return df_accidentes
```

# Centralizacion de Accidentes en la Provincia de Córdoba

```python
def accidentes_por_provincia(provincia, df):
    return df[df["provincia_nombre"] == provincia]
```

Debido a la complejidad en la predicción de diferentes tipos de clases a causas del balanceo del dataset en supervisado se opto por tomar solo 3 clases:
+ Colisión vehículo-objeto
+ Colisión vehículo-vehículo
+ Vuelco-Despistes

Dejando de lado los accidentes cuyo modo de producción son:

+ Otro
+ Sin determinar
+ Colisión vehículo-persona

```python
def accidentes_por_provincia2(provincia, df):
    df_nuevo = accidentes_por_provincia(provincia,df)
    atributo_salida = "modo_produccion_hecho"
    df_nuevo = df_nuevo[
        (df_nuevo[atributo_salida] != "Otro")
        & (df_nuevo[atributo_salida] != "Sin determinar")
        & (df_nuevo[atributo_salida] != "Colisión vehículo-persona")
    ].copy()
    return df_nuevo
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from kmodes.kmodes import KModes
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import silhouette_score, calinski_harabasz_score


def metodo_elbow_kmodes(df, lista_categoricos, max_clusters=10):
    """
    Aplica el método de Elbow para K-Modes.

    Parámetros:
    - df: DataFrame original
    - lista_categoricos: lista de variables categóricas
    - max_clusters: número máximo de clusters a evaluar

    Retorna:
    - Diccionario con costos de clustering
    """
    # Preprocesar datos
    df_processed, _ = preprocesar_datos_categoricos(df, lista_categoricos)

    # Almacenar costos
    costos = []
    k_valores = range(1, max_clusters + 1)

    # Calcular costos para diferentes números de clusters
    for k in k_valores:
        kmodes = KModes(n_clusters=k, init="Huang", n_init=5, random_state=42)
        kmodes.fit(df_processed[lista_categoricos])
        costos.append(kmodes.cost_)

    # Graficar método de Elbow
    plt.figure(figsize=(10, 5))
    plt.plot(k_valores, costos, "bx-")
    plt.xlabel("Número de clusters (k)")
    plt.ylabel("Costo de K-Modes")
    plt.title("Método de Elbow - K-Modes")
    plt.show()

    return {"k_valores": k_valores, "costos": costos}


def preprocesar_datos_categoricos(df, lista_categoricos):
    """
    Preprocesa datos categóricos para K-Modes usando LabelEncoder.

    Parámetros:
    - df: DataFrame original
    - lista_categoricos: lista de columnas categóricas

    Retorna:
    - DataFrame preprocesado
    - Lista de objetos de codificación
    """
    # Crear copia del DataFrame
    df_processed = df.copy()

    # Inicializar lista de encoders
    encoders = []

    # Codificar variables categóricas
    for columna in lista_categoricos:
        # Crear LabelEncoder para cada columna
        le = LabelEncoder()
        df_processed[columna] = le.fit_transform(df_processed[columna].astype(str))
        encoders.append(le)

    return df_processed, encoders


def evaluar_kmodes_clusters(df_processed, lista_categoricos, max_clusters=10):
    """
    Evalúa la calidad del clustering K-Modes para diferentes números de clusters.

    Parámetros:
    - df_processed: DataFrame preprocesado
    - lista_categoricos: lista de variables categóricas
    - max_clusters: número máximo de clusters a evaluar

    Retorna:
    - Diccionario con métricas de calidad para cada número de clusters
    """
    # Preparar diccionario para almacenar resultados
    metricas_clustering = {
        "n_clusters": [],
        "silhouette_score": [],
        "calinski_harabasz_score": [],
        "costo_kmodes": [],
    }

    # Evaluar clustering para diferentes números de clusters
    for n_clusters in range(2, max_clusters + 1):
        # Aplicar K-Modes
        kmodes = KModes(n_clusters=n_clusters, init="Huang", n_init=5, random_state=42)
        clusters = kmodes.fit_predict(df_processed[lista_categoricos])

        # Calcular métricas de calidad
        try:
            # Silhouette Score (con transformación a numérico)
            silhouette = silhouette_score(
                df_processed[lista_categoricos].apply(pd.to_numeric), clusters
            )

            # Calinski-Harabasz Score
            calinski = calinski_harabasz_score(
                df_processed[lista_categoricos].apply(pd.to_numeric), clusters
            )
        except Exception as e:
            print(f"Error calculando métricas para {n_clusters} clusters: {e}")
            silhouette = None
            calinski = None

        # Almacenar resultados
        metricas_clustering["n_clusters"].append(n_clusters)
        metricas_clustering["silhouette_score"].append(silhouette)
        metricas_clustering["calinski_harabasz_score"].append(calinski)
        metricas_clustering["costo_kmodes"].append(kmodes.cost_)

        print(f"Clusters: {n_clusters}")
        if silhouette is not None:
            print(f"Silhouette Score: {silhouette:.4f}")
            print(f"Calinski-Harabasz Score: {calinski:.4f}\n")

    # Visualizar resultados
    plt.figure(figsize=(15, 5))

    # Gráfico de Silhouette Score
    plt.subplot(1, 3, 1)
    plt.plot(
        metricas_clustering["n_clusters"],
        metricas_clustering["silhouette_score"],
        marker="o",
        color="green",
    )
    plt.title("Silhouette Score")
    plt.xlabel("Número de Clusters")
    plt.ylabel("Silhouette Score")

    # Gráfico de Calinski-Harabasz Score
    plt.subplot(1, 3, 2)
    plt.plot(
        metricas_clustering["n_clusters"],
        metricas_clustering["calinski_harabasz_score"],
        marker="o",
        color="red",
    )
    plt.title("Calinski-Harabasz Score")
    plt.xlabel("Número de Clusters")
    plt.ylabel("Calinski-Harabasz Score")

    plt.tight_layout()
    plt.show()

    return metricas_clustering


def seleccionar_mejor_numero_clusters(metricas):
    """
    Selecciona el mejor número de clusters basándose en las métricas.

    Estrategias:
    - Silhouette Score: Valor máximo
    - Calinski-Harabasz Score: Valor máximo
    """
    # Obtener índice del máximo Silhouette Score
    idx_silhouette = np.argmax(metricas["silhouette_score"])
    # Obtener índice del máximo Calinski-Harabasz Score
    idx_calinski = np.argmax(metricas["calinski_harabasz_score"])

    mejor_clusters_silhouette = metricas["n_clusters"][idx_silhouette]
    mejor_clusters_calinski = metricas["n_clusters"][idx_calinski]

    print("\nMejor número de clusters:")
    print(f"Según Silhouette Score: {mejor_clusters_silhouette}")
    print(f"Según Calinski-Harabasz Score: {mejor_clusters_calinski}")

    return mejor_clusters_silhouette


def aplicar_kmodes_clustering(df, lista_categoricos, n_clusters):
    """
    Aplica clustering K-Modes y agrega los clusters al DataFrame original.

    Parámetros:
    - df: DataFrame original
    - lista_categoricos: lista de variables categóricas
    - n_clusters: número de clusters

    Retorna:
    - DataFrame con columna de clusters
    - Modelo K-Modes
    - Lista de encoders
    """
    # Preprocesar datos
    df_processed, encoders = preprocesar_datos_categoricos(df, lista_categoricos)

    # Aplicar K-Modes
    kmodes = KModes(n_clusters=n_clusters, init="Huang", n_init=5, random_state=42)
    clusters = kmodes.fit_predict(df_processed[lista_categoricos])

    # Agregar columna de clusters al DataFrame original
    df_resultado = df.copy()
    df_resultado["Cluster"] = clusters

    return df_resultado, kmodes, encoders


def analizar_clusters_categoricos(df_resultado, lista_categoricos):
    """
    Analiza las características de los clusters categóricos.
    """
    # Análisis de variables categóricas
    print("\n=== Análisis de Variables Categóricas por Cluster ===")
    for cat in lista_categoricos:
        print(f"\nDistribución de {cat} por cluster:")
        tabla_distribucion = (
            pd.crosstab(df_resultado["Cluster"], df_resultado[cat], normalize="index")
            * 100
        )
        print(tabla_distribucion)

        # Visualización de la distribución
        plt.figure(figsize=(12, 6))
        tabla_distribucion.plot(kind="bar", stacked=True)
        plt.title(f"Distribución de {cat} por Cluster")
        plt.xlabel("Cluster")
        plt.ylabel("Porcentaje")
        plt.legend(title=cat, bbox_to_anchor=(1.05, 1), loc="upper left")
        plt.tight_layout()
        plt.show()


def ejecutar_analisis_kmodes(df, lista_categoricos):
    """
    Ejecuta análisis completo de clustering con K-Modes.
    """
    # Preprocesar datos
    df_processed, _ = preprocesar_datos_categoricos(df, lista_categoricos)

    # Evaluar clusters
    metricas = evaluar_kmodes_clusters(df_processed, lista_categoricos)

    # Seleccionar mejor número de clusters
    n_clusters = seleccionar_mejor_numero_clusters(metricas)

    # Aplicar clustering
    df_resultado, modelo, encoder = aplicar_kmodes_clustering(
        df, lista_categoricos, n_clusters
    )

    # Analizar resultados
    analizar_clusters_categoricos(df_resultado, lista_categoricos)

    return df_resultado, modelo, encoder
```

### Clustering con K-Prototypes

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from kmodes.kprototypes import KPrototypes
from sklearn.metrics import silhouette_score, calinski_harabasz_score


def metodo_elbow_kprototypes(df, lista_numericos, lista_categoricos, max_clusters=10):
    """
    Aplica el método de Elbow para K-Prototypes.

    Parámetros:
    - df: DataFrame original
    - lista_numericos: lista de variables numéricas
    - lista_categoricos: lista de variables categóricas
    - max_clusters: número máximo de clusters a evaluar

    Retorna:
    - Diccionario con costos de clustering
    """
    # Preprocesar datos
    df_processed, scaler = preprocesar_datos(df, lista_numericos, lista_categoricos)

    # Preparar índices de columnas categóricas
    categorical_indices = list(
        range(len(lista_numericos), len(lista_numericos) + len(lista_categoricos))
    )

    # Almacenar costos
    costos = []
    k_valores = range(1, max_clusters + 1)

    # Calcular costos para diferentes números de clusters
    for k in k_valores:
        kproto = KPrototypes(n_clusters=k, init="Huang", random_state=42)
        kproto.fit(
            df_processed[lista_numericos + lista_categoricos].values,
            categorical=categorical_indices,
        )
        costos.append(kproto.cost_)

    # Graficar método de Elbow
    plt.figure(figsize=(10, 5))
    plt.plot(k_valores, costos, "bx-")
    plt.xlabel("Número de clusters (k)")
    plt.ylabel("Costo de K-Prototypes")
    plt.title("Método de Elbow - K-Prototypes")
    plt.show()

    return {"k_valores": k_valores, "costos": costos}


def preprocesar_datos(df, lista_numericos, lista_categoricos):
    """
    Preprocesa los datos para clustering usando LabelEncoder para variables categóricas.

    Parámetros:
    - df: DataFrame original
    - lista_numericos: lista de variables numéricas
    - lista_categoricos: lista de variables categóricas

    Retorna:
    - DataFrame preprocesado
    - Diccionario de label encoders para cada variable categórica
    """
    # Crear copia del DataFrame
    df_processed = df.copy()

    # Almacenar label encoders
    label_encoders = {}

    # Aplicar LabelEncoder a variables categóricas
    for col in lista_categoricos:
        le = LabelEncoder()
        df_processed[col] = le.fit_transform(df_processed[col].astype(str))
        label_encoders[col] = le

    # Normalizar variables numéricas
    scaler = StandardScaler()
    df_processed[lista_numericos] = scaler.fit_transform(df_processed[lista_numericos])

    return df_processed, {"scaler": scaler, "label_encoders": label_encoders}


def evaluar_clusters(df_processed, lista_numericos, lista_categoricos, max_clusters=10):
    """
    Evalúa la calidad del clustering para diferentes números de clusters.

    Parámetros:
    - df_processed: DataFrame preprocesado
    - lista_numericos: lista de variables numéricas
    - lista_categoricos: lista de variables categóricas
    - max_clusters: número máximo de clusters a evaluar

    Retorna:
    - Diccionario con métricas de calidad para cada número de clusters
    """
    # Definir índices de variables categóricas para K-Prototypes
    categorical_indices = list(
        range(len(lista_numericos), len(lista_numericos) + len(lista_categoricos))
    )

    # Preparar diccionario para almacenar resultados
    metricas_clustering = {
        "n_clusters": [],
        "silhouette_score": [],
        "calinski_harabasz_score": [],
    }

    # Evaluar clustering para diferentes números de clusters
    for n_clusters in range(2, max_clusters + 1):
        # Aplicar K-Prototypes
        kproto = KPrototypes(n_clusters=n_clusters, init="Huang", random_state=42)

        # Preparar datos para clustering
        X = df_processed[lista_numericos + lista_categoricos].values

        # Obtener clusters
        clusters = kproto.fit_predict(X, categorical=categorical_indices)

        # Calcular métricas de calidad
        silhouette = silhouette_score(df_processed[lista_numericos], clusters)

        calinski = calinski_harabasz_score(df_processed[lista_numericos], clusters)

        # Almacenar resultados
        metricas_clustering["n_clusters"].append(n_clusters)
        metricas_clustering["silhouette_score"].append(silhouette)
        metricas_clustering["calinski_harabasz_score"].append(calinski)

        print(f"Clusters: {n_clusters}")
        print(f"Silhouette Score: {silhouette:.4f}")
        print(f"Calinski-Harabasz Score: {calinski:.4f}\n")

    # Visualizar resultados
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(
        metricas_clustering["n_clusters"],
        metricas_clustering["silhouette_score"],
        marker="o",
    )
    plt.title("Silhouette Score")
    plt.xlabel("Número de Clusters")
    plt.ylabel("Silhouette Score")

    plt.subplot(1, 2, 2)
    plt.plot(
        metricas_clustering["n_clusters"],
        metricas_clustering["calinski_harabasz_score"],
        marker="o",
        color="red",
    )
    plt.title("Calinski-Harabasz Score")
    plt.xlabel("Número de Clusters")
    plt.ylabel("Calinski-Harabasz Score")

    plt.tight_layout()
    plt.show()

    return metricas_clustering


def seleccionar_mejor_numero_clusters(metricas):
    """
    Selecciona el mejor número de clusters basándose en las métricas.

    Estrategias:
    - Silhouette Score: Valor máximo
    - Calinski-Harabasz Score: Valor máximo
    """
    # Obtener índice del máximo Silhouette Score
    idx_silhouette = np.argmax(metricas["silhouette_score"])
    # Obtener índice del máximo Calinski-Harabasz Score
    idx_calinski = np.argmax(metricas["calinski_harabasz_score"])

    mejor_clusters_silhouette = metricas["n_clusters"][idx_silhouette]
    mejor_clusters_calinski = metricas["n_clusters"][idx_calinski]

    print("\nMejor número de clusters:")
    print(f"Según Silhouette Score: {mejor_clusters_silhouette}")
    print(f"Según Calinski-Harabasz Score: {mejor_clusters_calinski}")

    return mejor_clusters_silhouette


def aplicar_clustering(
    df, df_processed, lista_numericos, lista_categoricos, n_clusters
):
    """
    Aplica K-Prototypes clustering con el número óptimo de clusters.

    Modificada para trabajar con el nuevo preprocesamiento.
    """
    categorical_indices = list(
        range(len(lista_numericos), len(lista_numericos) + len(lista_categoricos))
    )

    kproto = KPrototypes(n_clusters=n_clusters, init="Huang", random_state=42)
    clusters = kproto.fit_predict(
        df_processed[lista_numericos + lista_categoricos].values,
        categorical=categorical_indices,
    )

    # Agregar clusters al DataFrame original
    df_resultado = df.copy()
    df_resultado["Cluster"] = clusters

    return df_resultado, kproto


def analizar_clusters(
    df_resultado, lista_numericos, lista_categoricos, preprocessing_info
):
    """
    Analiza las características de cada cluster.

    Modificada para trabajar con el nuevo preprocesamiento.
    """
    # Desempaquetar información de preprocesamiento
    scaler = preprocessing_info["scaler"]
    label_encoders = preprocessing_info["label_encoders"]

    # Análisis de variables numéricas
    print("\n=== Análisis de Variables Numéricas por Cluster ===")
    medias = df_resultado.groupby("Cluster")[lista_numericos].mean()
    print("\nMedias por cluster:")
    print(medias)

    # Análisis de variables categóricas
    print("\n=== Análisis de Variables Categóricas por Cluster ===")
    for cat in lista_categoricos:
        print(f"\nDistribución de {cat} por cluster:")
        # Usar label encoder para decodificar los valores
        crosstab = (
            pd.crosstab(df_resultado["Cluster"], df_resultado[cat], normalize="index")
            * 100
        )
        print(crosstab)

        # Visualización
        import matplotlib.pyplot as plt

        plt.figure(figsize=(10, 5))
        crosstab.plot(kind="bar", stacked=True, figsize=(12, 6))
        plt.title(f"Distribución de {cat} por Cluster")
        plt.ylabel("Proporción")
        plt.xlabel("Cluster")
        plt.legend(title=cat, bbox_to_anchor=(1.05, 1), loc="upper left")
        plt.tight_layout()
        plt.show()

    # Visualización de atributos numéricos por cluster
    for col in lista_numericos:
        plt.figure(figsize=(10, 5))
        df_resultado.boxplot(column=col, by="Cluster", grid=False)
        plt.title(f"Distribución de {col} por Cluster")
        plt.suptitle("")  # Quitar el título de superposición generado automáticamente
        plt.xlabel("Cluster")
        plt.ylabel(col)
        plt.show()


def ejecutar_analisis_completo(df, lista_numericos, lista_categoricos):
    """
    Ejecuta el análisis completo de clustering con evaluación de métricas.

    Devuelve:
    - DataFrame con resultados de clustering
    - Modelo K-Prototypes
    - Información de preprocesamiento (scaler y label encoders)
    """
    # Preprocesamiento
    df_processed, preprocessing_info = preprocesar_datos(
        df, lista_numericos, lista_categoricos
    )

    # Evaluar clusters
    metricas = evaluar_clusters(df_processed, lista_numericos, lista_categoricos)

    # Seleccionar mejor número de clusters
    n_clusters = seleccionar_mejor_numero_clusters(metricas)

    # Aplicar clustering
    df_resultado, modelo = aplicar_clustering(
        df, df_processed, lista_numericos, lista_categoricos, n_clusters
    )

    # Analizar resultados
    analizar_clusters(
        df_resultado, lista_numericos, lista_categoricos, preprocessing_info
    )

    return df_resultado, modelo, preprocessing_info
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

# Técnicas de sobremuestreo
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.pipeline import Pipeline as ImbPipeline


def preparar_dataset(df, atributos_entrada, atributo_salida, categorical_columns):
    """
    Prepara el dataset para entrenar modelos

    Parámetros:
    - df: DataFrame original
    - atributos_entrada: Lista de características de entrada
    - atributo_salida: Variable objetivo
    - categorical_columns: Columnas categóricas a codificar

    Retorna:
    - X: Features
    - y: Variable objetivo
    - label_encoders: Diccionario de label encoders
    """
    # Seleccionar columnas
    df = df[atributos_entrada + [atributo_salida]]

    # Codificar variables categóricas
    label_encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le

    # Separar features y target
    X = df.drop(columns=[atributo_salida])
    y = df[atributo_salida]

    return X, y, label_encoders


def entrenar_y_evaluar_modelos(X, y, tecnicas_balanceo=None):
    """
    Entrena y evalúa modelos con diferentes estrategias de balanceo

    Parámetros:
    - X: Features
    - y: Variable objetivo
    - tecnicas_balanceo: Lista de técnicas de balanceo a probar

    Retorna:
    - Resultados de los modelos
    """
    # Dividir datos
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )

    # Configuración de modelos
    param_grid = {
        "n_estimators": [200, 300, 500],
        "max_features": ["sqrt", "log2"],
        "max_depth": [7, 9, 13, 15, 17],
        "criterion": ["gini", "entropy"],
    }

    # Resultados
    resultados = {}

    # Modelo base sin balanceo
    rf_base = RandomForestClassifier(random_state=42)
    rf_base.fit(X_train, y_train)
    y_pred_base = rf_base.predict(X_test)
    resultados["Base"] = {
        "accuracy": accuracy_score(y_test, y_pred_base),
        "clasificacion_report": classification_report(y_test, y_pred_base),
    }

    # Modelos con técnicas de balanceo
    if tecnicas_balanceo is None:
        tecnicas_balanceo = [
            ("SMOTE", SMOTE(random_state=42)),
            ("RandomOverSampler", RandomOverSampler(random_state=42)),
        ]

    for nombre, tecnica_balanceo in tecnicas_balanceo:
        # Pipeline con escalamiento, balanceo y modelo
        pipeline = ImbPipeline(
            [
                ("scaler", StandardScaler()),
                ("balanceo", tecnica_balanceo),
                ("clasificador", RandomForestClassifier(random_state=42)),
            ]
        )

        # Búsqueda de hiperparámetros
        grid_search = GridSearchCV(
            pipeline,
            param_grid={"clasificador__" + k: v for k, v in param_grid.items()},
            cv=4,
            scoring="f1_weighted",
        )

        grid_search.fit(X_train, y_train)

        # Predicción y evaluación
        y_pred = grid_search.predict(X_test)
        resultados[nombre] = {
            "accuracy": accuracy_score(y_test, y_pred),
            "clasificacion_report": classification_report(y_test, y_pred),
            "mejor_modelo": grid_search.best_estimator_,
            "mejores_parametros": grid_search.best_params_,
        }

    return resultados


def main():
    df = accidentes_por_provincia2("Córdoba", obtener_accidentes())

    # Definir columnas
    atributos_entrada = [
        "nro_inputados",
        "hora_hecho",
        "latitud",
        "longitud",
        "dia",
        "mes",
        "localidad_nombre",
    ]

    atributo_salida = "modo_produccion_hecho"

    categorical_columns = [
        "localidad_nombre",
    ]

    # Preparar dataset
    X, y, label_encoders = preparar_dataset(
        df, atributos_entrada, atributo_salida, categorical_columns
    )

    # Entrenar y evaluar modelos
    resultados = entrenar_y_evaluar_modelos(X, y)

    # Imprimir resultados
    for nombre, resultado in resultados.items():
        print(f"\nResultados para {nombre}:")
        print(f"Accuracy: {resultado['accuracy']}")
        print("Reporte de Clasificación:")
        print(resultado["clasificacion_report"])

        # Imprimir mejores parámetros si están disponibles
        if "mejores_parametros" in resultado:
            print("\nMejores Hiperparámetros:")
            for param, valor in resultado["mejores_parametros"].items():
                print(f"{param}: {valor}")

    return resultados


# Ejecutar el script
if __name__ == "__main__":
    main()


"""
Sin quitar Colisión vehículo-persona
Resultados para Base:
Accuracy: 0.7290969899665551
Reporte de Clasificación:
                            precision    recall  f1-score   support

  Colisión vehículo-objeto       0.64      0.64      0.64        45
 Colisión vehículo-persona       0.50      0.06      0.10        36
Colisión vehículo-vehículo       0.77      0.97      0.86       172
          Vuelco-Despistes       0.60      0.46      0.52        46

                  accuracy                           0.73       299
                 macro avg       0.63      0.53      0.53       299
              weighted avg       0.69      0.73      0.68       299


Resultados para SMOTE:
Accuracy: 0.6521739130434783
Reporte de Clasificación:
                            precision    recall  f1-score   support

  Colisión vehículo-objeto       0.68      0.58      0.63        45
 Colisión vehículo-persona       0.23      0.28      0.25        36
Colisión vehículo-vehículo       0.77      0.77      0.77       172
          Vuelco-Despistes       0.60      0.59      0.59        46

                  accuracy                           0.65       299
                 macro avg       0.57      0.55      0.56       299
              weighted avg       0.66      0.65      0.66       299


Resultados para RandomOverSampler:
Accuracy: 0.6521739130434783
Reporte de Clasificación:
                            precision    recall  f1-score   support

  Colisión vehículo-objeto       0.65      0.58      0.61        45
 Colisión vehículo-persona       0.23      0.25      0.24        36
Colisión vehículo-vehículo       0.77      0.80      0.78       172
          Vuelco-Despistes       0.55      0.48      0.51        46

                  accuracy                           0.65       299
                 macro avg       0.55      0.53      0.54       299
              weighted avg       0.65      0.65      0.65       299



Accuracy: 0.8479087452471483
Reporte de Clasificación:
                            precision    recall  f1-score   support

  Colisión vehículo-objeto       0.65      0.69      0.67        45
Colisión vehículo-vehículo       0.92      1.00      0.96       172
          Vuelco-Despistes       0.69      0.43      0.53        46

                  accuracy                           0.85       263
                 macro avg       0.75      0.71      0.72       263
              weighted avg       0.84      0.85      0.84       263


Resultados para SMOTE:
Accuracy: 0.8517110266159695
Reporte de Clasificación:
                            precision    recall  f1-score   support

  Colisión vehículo-objeto       0.70      0.62      0.66        45
Colisión vehículo-vehículo       0.92      0.99      0.96       172
          Vuelco-Despistes       0.66      0.54      0.60        46

                  accuracy                           0.85       263
                 macro avg       0.76      0.72      0.74       263
              weighted avg       0.84      0.85      0.84       263


Resultados para RandomOverSampler:
Accuracy: 0.8555133079847909
Reporte de Clasificación:
                            precision    recall  f1-score   support

  Colisión vehículo-objeto       0.70      0.67      0.68        45
Colisión vehículo-vehículo       0.92      0.99      0.96       172
          Vuelco-Despistes       0.69      0.52      0.59        46

                  accuracy                           0.86       263
                 macro avg       0.77      0.73      0.74       263
              weighted avg       0.84      0.86      0.85       263

"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.pipeline import Pipeline

# Importaciones de XGBoost
import xgboost as xgb
from xgboost import XGBClassifier

# Técnicas de sobremuestreo
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.pipeline import Pipeline as ImbPipeline


def preparar_dataset(df, atributos_entrada, atributo_salida, categorical_columns):
    """
    Prepara el dataset para entrenar modelos

    Parámetros:
    - df: DataFrame original
    - atributos_entrada: Lista de características de entrada
    - atributo_salida: Variable objetivo
    - categorical_columns: Columnas categóricas a codificar

    Retorna:
    - X: Features
    - y: Variable objetivo
    - label_encoders: Diccionario de label encoders
    """
    # Seleccionar columnas
    df = df[atributos_entrada + [atributo_salida]]

    # Codificar variables categóricas
    label_encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le

    # Separar features y target
    X = df.drop(columns=[atributo_salida])
    y = df[atributo_salida]

    return X, y, label_encoders


def entrenar_y_evaluar_modelos(X, y, tecnicas_balanceo=None):
    """
    Entrena y evalúa modelos XGBoost con diferentes estrategias de balanceo

    Parámetros:
    - X: Features
    - y: Variable objetivo
    - tecnicas_balanceo: Lista de técnicas de balanceo a probar

    Retorna:
    - Resultados de los modelos
    """
    # Dividir datos
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )

    # Configuración de hiperparámetros para XGBoost
    param_grid = {
        "max_depth": [3, 5, 7, 9],
        "learning_rate": [0.01, 0.1, 0.3],
        "n_estimators": [100, 300, 500],
        "min_child_weight": [1, 3, 5],
        "subsample": [0.6, 0.8, 1.0],
        "colsample_bytree": [0.6, 0.8, 1.0],
        "gamma": [0, 0.1, 0.3],
    }

    # Resultados
    resultados = {}

    # Modelo base sin balanceo
    xgb_base = XGBClassifier(
        use_label_encoder=False, eval_metric="mlogloss", random_state=42
    )
    xgb_base.fit(X_train, y_train)
    y_pred_base = xgb_base.predict(X_test)
    resultados["Base"] = {
        "accuracy": accuracy_score(y_test, y_pred_base),
        "clasificacion_report": classification_report(y_test, y_pred_base),
    }

    # Modelos con técnicas de balanceo
    if tecnicas_balanceo is None:
        tecnicas_balanceo = [
            ("SMOTE", SMOTE(random_state=42)),
            ("RandomOverSampler", RandomOverSampler(random_state=42)),
        ]

    for nombre, tecnica_balanceo in tecnicas_balanceo:
        # Pipeline con escalamiento, balanceo y modelo
        pipeline = ImbPipeline(
            [
                ("scaler", StandardScaler()),
                ("balanceo", tecnica_balanceo),
                (
                    "clasificador",
                    XGBClassifier(
                        use_label_encoder=False, eval_metric="mlogloss", random_state=42
                    ),
                ),
            ]
        )

        # Búsqueda de hiperparámetros con GridSearchCV
        grid_search = GridSearchCV(
            pipeline,
            param_grid={"clasificador__" + k: v for k, v in param_grid.items()},
            cv=4,
            scoring="f1_weighted",
            n_jobs=-1,  # Utiliza todos los núcleos disponibles
        )

        grid_search.fit(X_train, y_train)

        # Predicción y evaluación
        y_pred = grid_search.predict(X_test)
        resultados[nombre] = {
            "accuracy": accuracy_score(y_test, y_pred),
            "clasificacion_report": classification_report(y_test, y_pred),
            "mejor_modelo": grid_search.best_estimator_,
            "mejores_parametros": grid_search.best_params_,
        }

    return resultados


def main():
    df = accidentes_por_provincia2("Córdoba", obtener_accidentes())

    # Definir columnas
    atributos_entrada = [
        "nro_inputados",
        "hora_hecho",
        "latitud",
        "longitud",
        "dia",
        "mes",
        "localidad_nombre",
    ]

    atributo_salida = "modo_produccion_hecho"

    categorical_columns = ["localidad_nombre", "modo_produccion_hecho"]

    # Preparar dataset
    X, y, label_encoders = preparar_dataset(
        df, atributos_entrada, atributo_salida, categorical_columns
    )

    # Entrenar y evaluar modelos
    resultados = entrenar_y_evaluar_modelos(X, y)

    # Imprimir resultados
    for nombre, resultado in resultados.items():
        print(f"\nResultados para {nombre}:")
        print(f"Accuracy: {resultado['accuracy']}")
        print("Reporte de Clasificación:")
        print(resultado["clasificacion_report"])

        # Imprimir mejores parámetros si están disponibles
        if "mejores_parametros" in resultado:
            print("\nMejores Hiperparámetros:")
            for param, valor in resultado["mejores_parametros"].items():
                print(f"{param}: {valor}")

    return resultados


if __name__ == "__main__":
    main()


"""
Sin quitar Colisión vehículo-persona
Resultados para Base:
Accuracy: 0.68561872909699
Reporte de Clasificación:
              precision    recall  f1-score   support

           0       0.67      0.62      0.64        45
           1       0.15      0.08      0.11        36
           2       0.76      0.88      0.82       172
           3       0.61      0.48      0.54        46

    accuracy                           0.69       299
   macro avg       0.55      0.52      0.53       299
weighted avg       0.65      0.69      0.66       299


Resultados para SMOTE:
Accuracy: 0.6354515050167224
Reporte de Clasificación:
              precision    recall  f1-score   support

           0       0.61      0.49      0.54        45
           1       0.29      0.44      0.35        36
           2       0.79      0.76      0.77       172
           3       0.50      0.48      0.49        46

    accuracy                           0.64       299
   macro avg       0.55      0.54      0.54       299
weighted avg       0.66      0.64      0.64       299


Mejores Hiperparámetros:
clasificador__colsample_bytree: 0.6
clasificador__gamma: 0.3
clasificador__learning_rate: 0.01
clasificador__max_depth: 9
clasificador__min_child_weight: 5
clasificador__n_estimators: 100
clasificador__subsample: 0.6

Resultados para RandomOverSampler:
Accuracy: 0.6822742474916388
Reporte de Clasificación:
              precision    recall  f1-score   support

           0       0.64      0.62      0.63        45
           1       0.22      0.14      0.17        36
           2       0.77      0.87      0.82       172
           3       0.57      0.46      0.51        46

    accuracy                           0.68       299
   macro avg       0.55      0.52      0.53       299
weighted avg       0.65      0.68      0.66       299


Mejores Hiperparámetros:
clasificador__colsample_bytree: 0.6
clasificador__gamma: 0.1
clasificador__learning_rate: 0.1
clasificador__max_depth: 9
clasificador__min_child_weight: 1
clasificador__n_estimators: 500
clasificador__subsample: 0.8

MINUTOS DE EJECUCION 54 MINUTOS
"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
from imblearn.pipeline import Pipeline as ImbPipeline


def preparar_dataset(df, atributos_entrada, atributo_salida, categorical_columns):
    """
    Prepara el dataset para entrenar modelos
    """
    # Seleccionar columnas
    df = df[atributos_entrada + [atributo_salida]]

    # Codificar variables categóricas de entrada
    label_encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le

    # Codificar variable objetivo
    le_target = LabelEncoder()
    df[atributo_salida] = le_target.fit_transform(df[atributo_salida])

    # Separar features y target
    X = df.drop(columns=[atributo_salida])
    y = df[atributo_salida]

    return X, y, label_encoders, le_target


def generar_modelo_xgboost_cv(df):
    # Definir columnas
    atributos_entrada = [
        "nro_inputados",
        "hora_hecho",
        "latitud",
        "longitud",
        "dia",
        "mes",
        "localidad_nombre",
    ]
    atributo_salida = "modo_produccion_hecho"
    categorical_columns = ["localidad_nombre"]

    # Preparar dataset
    X, y, label_encoders, le_target = preparar_dataset(
        df, atributos_entrada, atributo_salida, categorical_columns
    )

    # Dividir datos
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )

    # Pipeline con SMOTE y Tomek Links
    modelo = ImbPipeline(
        [
            ("scaler", StandardScaler()),
            ("smote", SMOTE(random_state=42)),  # Sobremuestreo de la clase minoritaria
            ("tomek", TomekLinks()),  # Eliminación de enlaces de Tomek
            (
                "classifier",
                XGBClassifier(
                    colsample_bytree=1.0,
                    gamma=0.1,
                    learning_rate=0.01,
                    max_depth=5,
                    min_child_weight=1,
                    n_estimators=100,
                    subsample=0.8,
                    random_state=42,
                    use_label_encoder=False,
                    eval_metric="mlogloss",
                ),
            ),
        ]
    )

    # Realizar Cross-Validation
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(modelo, X, y, cv=cv, scoring="accuracy")

    print("Resultados de Cross-Validation:")
    print(f"Precisión CV: {cv_scores}")
    print(f"Precisión Media CV: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")

    # Entrenar modelo
    modelo.fit(X_train, y_train)

    # Predicciones
    pred_train = modelo.predict(X_train)
    pred_test = modelo.predict(X_test)

    # Métricas de evaluación
    train_acc = accuracy_score(y_train, pred_train)
    test_acc = accuracy_score(y_test, pred_test)
    print(f"\nTrain Accuracy: {train_acc:.2f}")
    print(f"Test Accuracy: {test_acc:.2f}")

    print("\nReporte de Clasificación:")
    print(classification_report(y_test, pred_test, target_names=le_target.classes_))

    # Matriz de confusión
    cm = confusion_matrix(y_test, pred_test)

    # Visualizar matriz de confusión
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=le_target.classes_,
        yticklabels=le_target.classes_,
    )
    plt.title("Matriz de Confusión - XGBoost con SMOTE y Tomek Links")
    plt.xlabel("Predicción")
    plt.ylabel("Valor Real")
    plt.tight_layout()
    plt.show()

    return cm, cv_scores, le_target


def main():
    df = accidentes_por_provincia2("Córdoba", obtener_accidentes())
    cm, cv_scores, le_target = generar_modelo_xgboost_cv(df)

    print("\nMatriz de Confusión:")
    print(cm)

    return cm, cv_scores, le_target


if __name__ == "__main__":
    main()
```
